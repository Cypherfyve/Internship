**Q1. Days (x-axis) represents 8 days and Speed represents a carâ€™s speed. Plot a Basic line/curve plot between days and car speed, put x axis label as days and y axis label as car speed and put title Car Speed Measurement.**<br>
         Days=[1,2,3,4,5,6,7,8]<br>
         Speed=[60,62,61,58,56,57,46,63]

```python
import matplotlib.pyplot as plt
Days = [1,2,3,4,5,6,7,8]
Speed = [60,62,61,58,56,57,46,63]
plt.plot(Days, Speed, color = 'blue') # For Create a Line Plot
plt.xlabel('Days')
plt.ylabel('Car Speed')
plt.title('Car Speed Measurement')
plt.show()
```

**Q2. Now to above car data apply some string formats  like line style example green dotted line, marker shape like +, change markersize, markerface color etc.**

```python
import matplotlib.pyplot as plt
Days = [1,2,3,4,5,6,7,8]
Speed = [60,62,61,58,56,57,46,63]
plt.plot(Days, Speed, color = 'blue', marker = 'o', ms = 20, mec = 'red', mfc = 'green')
# ms for marker size, mfc for marker face color, mec for border color
plt.xlabel('Days')
plt.ylabel('Car Speed')
plt.title('Car Speed Measurement')
plt.show()
```

**Q3. Plot Axes Labels, Chart title, Legend in Car minimum, Maximum and average speed in 8 days as below.**<br>
max_speed=[80,91,92,88,77,79,76,75]<br>
min_speed=[42,43,40,42,33,36,34,35]<br>
avg_speed=[46,58,57,56,40,42,41,36]<br>
**Note: Utilize subplots()**

```python
import matplotlib.pyplot as plt
days = [1,2,3,4,5,6,7,8]
max_speed = [80,91,92,88,77,79,76,75]
min_speed = [42,43,40,42,33,36,34,35]
avg_speed = [46,58,57,56,40,42,41,36]
#plt.figure(figsize=(8,6))
fig, axes = plt.subplots(3,1,figsize=(8,6))
axes[0].plot(Days, max_speed, color = 'red', marker = '*', ms = 10, mec = 'red', mfc='red')
axes[0].set_xlabel('Days')
axes[0].set_ylabel('Maximum Speed')
axes[0].set_title('Car Speed Measurement')
axes[0].legend()
axes[1].plot(Days, min_speed, color = 'green', marker = 'o', ms = 10, mec = 'black')
axes[1].set_xlabel('Days')
axes[1].set_ylabel('Minimum Speed')
axes[1].set_title('Car Speed Measurement')
axes[1].legend()
axes[2].plot(Days, avg_speed, color = 'yellow', marker = '^', ms = 10, mec = 'red')
axes[2].set_xlabel('Days')
axes[2].set_ylabel('Average Speed')
axes[2].set_title('Car Speed Measurement')
axes[2].legend()
axes[2].grid()
plt.tight_layout()
```

**Q4. Plotting a basic sine graph by adding more features. Also adding Multiple plots by Superimposition like cosine wave.**

```python
import numpy as np
x = np.arange(0,4*np.pi,0.1)
y = np.sin(x)
z = np.cos(x)
plt.plot(x,y, color = 'green', marker='o')
plt.plot(x,z, color = 'blue', marker = '*')
plt.xlabel('Sin and Cos Value From 0 to 4pi')
plt.ylabel('Sin(x) and Cos(x)')
plt.title('Sin And Cosine Wave')
plt.legend(['sin(X)','Cos(x)'])
plt.show()
```
